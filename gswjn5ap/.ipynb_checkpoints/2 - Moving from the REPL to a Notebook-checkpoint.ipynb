```python
text = 'the quick brown fox jumped over the lazy dog'
counter = Counter()
for word in text.split(' '):
    counter[word[0]] += 1
counter['t']
```

```python
# One solution: CSV module imported from Python standard library
# Works but the CSV module is not concerned with display

import csv

file = open('data/weather.csv', 'r')
rdr = csv.reader(file)
for line in rdr:
    print(line)
file.close()
```

```python
# Instead we will use Pandas - a library constructed specifically for data analysis
# converts a csv file to a dataframe like in R

import pandas as pd

df_weather = pd.read_csv('data/weather.csv')

df_weather

# Notice the rich formatting
```

```python
# Lets clean up this df a little bit:

df_weather.info()

df_weather = df_weather.drop(df_weather.iloc[:, 5:7], axis = 1)

print('---------------------- \n')

df_weather.info()
```

```python
# Format column types

df_weather["Date"] = pd.to_datetime(df_weather["Date"])

df_weather.info()
```

```python
import numpy as np

df_weather['Precip'] = pd.to_numeric(df_weather['Precip'], errors='coerce')

df_weather = df_weather.astype({'Max Temp':'float','Min Temp':'float', 'New Snow':'float', 'Snow Depth':'float'}, errors='ignore')

df_weather.info()
```

```python
df_weather
```

```python
# The above table is rendered using HTML, therefore the style can be modified via the style property of the df
# df_weather.style.render()

# For example, let's highlight the max value in each column
df_weather.style.highlight_max()
```

```python
# Other examples of custom highlighting - any value that is 70 or greater
def highlight_70_and_over(precip):
    if isinstance(precip, float):
        return 'color: {0}'.format('green' if precip >= 70 else 'black')
    return 'color: black'
```

```python
df_weather.style.applymap(highlight_70_and_over)
```

```python
# Exporting formatted weather data
df_weather.to_csv('data/weather_formatted.csv', index = False)
```
