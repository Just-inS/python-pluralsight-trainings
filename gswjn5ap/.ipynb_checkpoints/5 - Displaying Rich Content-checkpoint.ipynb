
````python
# Importing numpy to generate some random data
import numpy as np

# Importing X to select items from a normal random distribution
np.random.seed(15976)
SIZE = 10000
x = np.random.normal(size = SIZE)
x
````

````python
plt.hist(x, bins = 50)
````

````python
xu = np.random.uniform(size = SIZE)
plt.hist(xu)
````

````python
y = np.random.normal(size = SIZE)
yu = np.random.uniform(size = SIZE)
````

````python
plt.scatter(x, y)
````

````python
plt.scatter(xu, yu)
````

````python
# Visualizing our Weather Data
import pandas as pd
weather = pd.read_csv('data/weather_formatted.csv')
weather.head(6)
````

````python
# Isolating Precipitation
precip = weather.iloc[:, [0,5]]
precip.head()
````

````python
type(precip)
````

````python
# Get max and min values
plt.bar(precip.iloc[:,0], precip.iloc[:,1])
plt.xticks(rotation = 90)
plt.xlabel("Date")
plt.ylabel("Precip (in)")
plt.title("Rainfall in Pittsburgh")
plt.savefig('data/rainfall_plot.png', bbox_inches='tight')
plt.show()
````
